### Aggregate Data with the Python .agg() Function

Let's work out the number of different themes shipped by year. This means we have to count the number of unique theme_ids per calendar year.

```python
themes = boxes.groupby("year").agg({"theme_id": pd.Series.nunique})
# .agg() method takes dictionary as an argument
```

```python
themes.rename(columns = {"theme_id": "nr_themes"}, inplace = True)
themes.head()
```

```python
themes.tail()
```

**Challenge**: Plot the number of themes released by year on a line chart. Only include the full calendar years (i.e., exclude 2020 and 2021).

```python
plt.figure(figsize=(10, 6))
plt.plot(themes.index[:-2], themes["nr_themes"][:-2], marker='o', linestyle='-')

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Number of Themes")
plt.title("Number of LEGO Themes Released Year-on-Year")

# Show the plot
plt.show()
```

### Line Charts with Two Separate Axes

```python
axis1 = plt.gca() # get current axis
axis2 = axis1.twinx() # create a second axis; twinx means they will share the same x-axis

axis1.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2], color = "g")
axis2.plot(themes.index[:-2], themes.nr_themes[:-2], "b")

# Add labels and title
axis1.set_xlabel("Year")
axis1.set_ylabel("Number of Themes", color = "green")
axis2.set_ylabel("Number of LEGO Themes Released Year-on-Year", color = "blue")
```

**Challenge**: Use the <code>.groupby()</code> and <code>.agg()</code> function together to figure out the average number of parts per set. How many parts did the average LEGO set released in 1954 compared to say, 2017?

### Database Schemas, Foreign Keys and Merging DataFrames

The themes.csv file has the actual theme names. The sets .csv has <code>theme_ids</code> which link to the <code>id</code> column in the themes.csv.

```python
# DATABASE SCHEMA = how the database is organized
# with some relational databases, like lego data, we have individual tables to split:
# the colors, the sets, and the themes

# think!!!
# sets.csv has "theme_ids", and this matches that the themes.csv has "id"
# so, "theme_ids" is the foreign key of sets.csv

# and, inside themes.csv, the "id" inside is unique
# so, "id" is the primary key inside themes.csv
```

### Merging (i.e., Combining) DataFrames based on a Key


```python
# let's merge 2 dataframes into one!! we're gonna combine theme_ids AND number of sets per theme
setThemes_count = boxes["theme_id"].value_counts()
setThemes_count[:5]
```

```python
setThemes_count = pd.DataFrame({"theme_id": setThemes_count.index, "set_count": setThemes_count.values})
setThemes_count[:5]
# there! :) much more organized
# note that we use a dictionary here (keys are in quotes and their values follow the colons) to make the dataframe
# also note: the dictionary KEYS become the column NAMES
```

```python
print(setThemes_count.columns)
print(legoThemes.columns)
# need to check where the column name discrepancy is
```

```python
# to merge 2 dataframes along ONE column, we need to provide BOTH dataframes and THEN column name to merge
# BOTH our setThemes_count and legoThemes dataframes have a column called "theme_id" so that is what we use

legoThemes = legoThemes.rename(columns={"id": "theme_id"})

mergedDF = pd.merge(setThemes_count, legoThemes, on="theme_id")
mergedDF[:3]
# this works, and theme_id is in my opinion more specific :)
```
